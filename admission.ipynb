# Storing Feature Matrix in A and Target in B 

```python
A = df.drop('Chance of Admit ', axis = 1)
```

```python
# independent variable
A
```

```python
# target/response variable
B = df['Chance of Admit ']
B
```

# Splitting Dataset into Training set and Test set

```python
from sklearn.model_selection import train_test_split
```

```python
A_train, A_test, B_train, B_test = train_test_split(A,B, test_size = 0.20, random_state = 42)
```

```python
A_train
```

```python
B_train #chances of admission
```

# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
A_train = sc.fit_transform(A_train)
A_test = sc.transform(A_test)
```

```python
A_train
```

# Applying ML Algorithms

```python
df.head(5)
```

Here, Our target variable is continous . So , we can impliment this as a regression problem

# Importing Regression Models

```python
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
```

```python
# Model Training

l = LinearRegression()
l.fit(A_train, B_train)

s = SVR()
s.fit(A_train, B_train)

r = RandomForestRegressor()
r.fit(A_train, B_train)

g = GradientBoostingRegressor()
g.fit(A_train, B_train)
```

```python
# prediction
B_predict_1 = l.predict(A_test)
B_predict_2 = s.predict(A_test)
B_predict_3 = r.predict(A_test)
B_predict_4 = g.predict(A_test)
```

# Evaluating the algorithm

```python
from sklearn import metrics
```

```python
score_1 = metrics.r2_score(B_test,B_predict_1)
score_2 = metrics.r2_score(B_test,B_predict_2)
score_3 = metrics.r2_score(B_test,B_predict_3)
score_4 = metrics.r2_score(B_test,B_predict_4)

print(score_1,score_2,score_3,score_4)
```

```python
final_value = pd.DataFrame({"Models" : ["LinearRegression", "SVR", "RandomForestRegressor", "GradientBoostingRegressor"],"R2_Score" : [score_1,score_2,score_3,score_4]})
```

```python
final_value
```

```python
# lets visualize
import seaborn as sns
```

```python
from matplotlib import pyplot as plt
plt.figure(figsize =(10,5))
sns.barplot(final_value,x= "Models" ,y = "R2_Score", palette= 'Blues' , width= 0.4)
```

Here, we can see that Linear Regression is the best model.

# Trying it as a Classification problem

```python
df.head(4)
```

For classification models, we have to convert our target variable , 'chance of admit' into a catagorical variable. 
I am going to convert it into zero or one using the threshold as 0.8. So, if the value is grater than 0.8 , i am taking 1 and otherwise 0.

```python
import numpy as np
```

```python
# list comprehension
B_train = [1 if value> 0.8 else 0 for value in B_train]
B_test = [1 if value> 0.8 else 0 for value in B_test]
```

```python
B_train = np.array(B_train)
B_test = np.array(B_test)
```

```python
B_train
```

# Importing Classification Models

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score #for evaluating
```

```python
# model training and evaluating
lr = LogisticRegression()
lr.fit(A_train, B_train)
B_predict_1 = lr.predict(A_test)
print(accuracy_score(B_test,B_predict_1))
```

```python
svm = SVC()
svm.fit(A_train, B_train)
B_predict_2 = svm.predict(A_test)
print(accuracy_score(B_test,B_predict_2))
```

```python
knc = KNeighborsClassifier()
knc.fit(A_train,B_train)
B_predict_3 = knc.predict(A_test)
print(accuracy_score(B_test,B_predict_3))
```

```python
rfc = RandomForestClassifier()
rfc.fit(A_train,B_train)
B_predict_4 = rfc.predict(A_test)
print(accuracy_score(B_test,B_predict_4))
```

```python
gbc =GradientBoostingClassifier()
gbc.fit(A_train,B_train)
B_predict_5 = gbc.predict(A_test)
print(accuracy_score(B_test,B_predict_5))
```

```python
# visualize accuracy as dataframe
final_value = pd.DataFrame({"Models" : ["LogisticRegression", "SVC", "KNeighborsClassifier", "RandomForestClassifier","GradientBoostingClassifier"],"accuracy_Score" : [accuracy_score(B_test,B_predict_1),
accuracy_score(B_test,B_predict_2),accuracy_score(B_test,B_predict_3),accuracy_score(B_test,B_predict_4),accuracy_score(B_test,B_predict_5)]} )
```

```python
final_value
```

```python
plt.figure(figsize =(15,5))
sns.barplot(final_value,x= "Models" ,y = "accuracy_Score", palette= 'Greens' , width= 0.4) # visualize
```

I am using GradientBoostingClassifier for production
